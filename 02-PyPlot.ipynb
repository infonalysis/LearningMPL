# MatplotLib Cheat Sheet

## Initialization

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

https://matplotlib.org/stable/api/pyplot_summary.html

```python
# create some data
data1, data2, data3, data4 = np.random.randn(4, 100)
names = ['apples', 'oranges', 'berries']
counts = [10, 5, 7]
data_dict = {
    'col1':data1,
    'col2':data2
}
pandas_df = pd.DataFrame(data_dict)
```

## Plotting

### plot(*args, scalex=True, scaley=True, data=None, **kwargs)
Plot y versus x as lines and/or markers.  
https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot  
- The optional parameter fmt is a convenient way for defining basic formatting like color, marker and linestyle.
- You can use Line2D properties as keyword arguments for more control on the appearance. Line properties and fmt can be mixed.
- When conflicting with fmt, keyword arguments take precedence.
- Line2D: https://matplotlib.org/stable/api/_as_gen/matplotlib.lines.Line2D.html#matplotlib.lines.Line2D


```python
# plotting a single list of values (x-axis is index)
plt.plot(data1)
```

```python
# plotting x-list and y-list with kwarg Line2D param
plt.plot(data1, data2, linewidth=2)
```

```python
# plotting x-list and y-list with named data
plt.plot('col1', 'col2', data=data_dict, linewidth=2)
```

```python
# plotting with a fmt specified line type
plt.plot(data1, data2, 'b--')
```

```python
# plotting multiple lists of values together, combining fmt and kwargs
plt.plot(data1, '--', color='green')
plt.plot(data2, '-', color='red')
plt.plot(data3, 'x', color='blue')
```

```python
# plotting categorical values
plt.plot(names, counts)
```

### text(x, y, s, fontdict=None, **kwargs)
Add text to the Axes.  
https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.text.html#matplotlib.pyplot.text

### annotate(text, xy, *args, **kwargs)
Annotate the point xy with text text. In the simplest form, the text is placed at xy. Optionally, the text can be displayed in another position xytext. An arrow pointing from the text to the annotated point xy can then be added by defining arrowprops.  
https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.annotate.html#matplotlib.pyplot.annotate

### grid(visible=None, which='major', axis='both', **kwargs)

## Multiple plots

### figure(num=None, figsize=None, dpi=None, facecolor=None, edgecolor=None, frameon=True, FigureClass=<class 'matplotlib.figure.Figure'>, clear=False, **kwargs)

### subplot(*args, **kwargs)
Add an Axes to the current figure or retrieve an existing Axes. This is a wrapper of Figure.add_subplot which provides additional behavior when working with the implicit API (see the notes section).  
https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplot.html#matplotlib.pyplot.subplot
