## Data structure hierarchy

* **Figure**
 * **Canvas**: The object that actually does the drawing to get you your plot.  
 * **Artists** (titles, figure legends, others???)
 * **Axes**: Each axes object represents what you think of as 'a plot'.  
   * **Artists** (which?)  
   * **Axis**: Each axis object is similar to a number line.  
     * **Artists** (which?): Everything visible on the figure is an artist  
     * **Locator**: Determines location of ticks
     * **Formatter**: Formats tick labels

## Basic plotting

### Helper function

```python
# Taken from https://matplotlib.org/stable/tutorials/introductory/usage.html#sphx-glr-tutorials-introductory-usage-py
def my_plotter(ax, data1, data2, param_dict):
    """
    A helper function to make a graph

    Parameters
    ----------
    ax : Axes
        The axes to draw to

    data1 : array
       The x data

    data2 : array
       The y data

    param_dict : dict
       Dictionary of keyword arguments to pass to ax.plot

    Returns
    -------
    out_plot : list
        list of artists added
    """
    out_plot = ax.plot(data1, data2, **param_dict)
    return out_plot
```

```python
# use plotter function
data1, data2, data3, data4 = np.random.randn(4, 100)
fig, ax = plt.subplots(1, 1)
my_plotter(ax, data1, data2, {'marker': 'x'})
```

```python
# use plotter function for multiple subplots
fig, (ax1, ax2) = plt.subplots(1, 2)
my_plotter(ax1, data1, data2, {'marker': 'x'})
my_plotter(ax2, data3, data4, {'marker': 'o'})
```
